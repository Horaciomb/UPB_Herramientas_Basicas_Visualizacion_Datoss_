¿Cómo se distribuye la variable "quality" en términos de frecuencia?

```python
df['quality'].value_counts().sort_index().plot(kind='bar')
```

```python
# definimos tamaño
plt.figure(figsize=(10,8))
# sns.set_theme(style="darkgrid")
plt.hist(df['quality'],color='#4f7358', alpha=0.7,edgecolor="black")
plt.title("Distribucion de calidad del vino")
plt.xlabel("Calidad")
plt.ylabel("Frecuencia")
plt.tight_layout()
plt.savefig("histograma.png",dpi=600)
plt.savefig("histograma.pdf")
plt.show()
```

¿Cuál es la relación entre la variable "alcohol" y la variable objetivo "quality"?

```python
# que tal con sns?
sns.set_theme(style="whitegrid")
#sns.color_palette("dark:#5A9_r", as_cmap=True)
sns.boxplot(x='quality', y='alcohol', data=df, palette= 'viridis')
plt.show()
```

### ¿Hay alguna correlación entre las variables "fixed acidity" y "pH"?

```python
# ###
plt.scatter(df['fixed acidity'], df['pH'],alpha=0.35,edgecolors="black",s=70)
plt.xlabel("Fixed axidity")
plt.ylabel("pH")
plt.title("Fixed acidity vs pH")
plt.show()
```

### ¿Cuál es el promedio de "fixed acidity" para cada calidad de vino?

```python
### mediante boxplot
sns.boxplot(data=df,x="quality",y="fixed acidity")
plt.title("Boxplot de la calidad de vino y acidez")
plt.show()
```

```python
### box plot y violin plot
sns.violinplot(data=df,x="quality",y="fixed acidity",palette='flare',legend=False,hue="quality")
plt.title("Boxplot de la calidad de vino y acidez")
plt.show()
```

### ¿Cuál es la media de "alcohol" para cada calidad de vino rojo?

```python
# Crear un gráfico de caja utilizando sns.boxplot
filtro=(df['red']==1)
df_vinos_rojos=df[filtro]
sns.boxplot(data=df_vinos_rojos,x='quality',y='alcohol',palette='flare',hue='quality',legend=False)
plt.show()
```

### ¿Cómo se distribuye la variable "citric acid" para los vinos con una calidad mayor o igual a 7?

```python
## en un hist plot
filtro2=df['quality']>=7
df_vinos_buenos=df[filtro2]
sns.histplot(data=df_vinos_buenos,x='citric acid',color='red')
plt.title("Distribución de Citric acid para los vinos buenos arriba de 7 en calidad")
plt.ylabel("cantidad")
plt.show()
```

## ¿Cuál es la relación entre "volatile acidity" y "quality" para los vinos con un pH mayor a 3?

```python
## scatter plot ?
filtro3=df['pH']>3
df_pH_mayor_3=df[filtro3]
sns.scatterplot(data=df_pH_mayor_3,x='volatile acidity',y='quality')
plt.show()
```

```python
# Crear un gráfico de caja utilizando sns.boxplot
sns.boxplot(data=df_pH_mayor_3,x='quality',y='volatile acidity',palette='flare',hue='quality',legend=False)
plt.show()
```
